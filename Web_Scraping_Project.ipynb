```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
name_info=[]
price_info=[]
rating_info=[]
```

```python
for i in range(1,3):
    if(i==1):
        url="https://www.flipkart.com/search?q=apple&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
    else:
        url="https://www.flipkart.com/search?q=apple&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"+"&page="+str(i)
    furl=requests.get(url)
    jsoup=BeautifulSoup(furl.content,'html.parser')
    products=jsoup.find_all('div',class_="_2kHMtA")
```

```python
for product in products:
    name=product.find('div',class_="_4rR01T").text.replace('\n','')
    price=product.find('div',class_="_30jeq3 _1_WHN1").text.replace('\n','')
    try:
        rating=product.find('div',class_="_3LWZlK").text.replace('\n','')
    except:
        rating="NULL"
    name_info.append(name)
    price_info.append(price)
    rating_info.append(rating)
```

```python
dict = {
    'Product Name':name_info,
    'Product Price':price_info,
    'Product Rating':rating_info
}
```

```python
df = pd.DataFrame(dict)
df
```

```python
df.to_csv("Apple_Products.csv",index=False)
```
